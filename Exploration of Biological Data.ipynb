```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
import seaborn as sns
from sklearn.linear_model import LinearRegression

df = pd.read_excel (r'C:\Users\Suyash\Documents\Work\COMP3212 - Computational Biology\Labs\Lab 1\Cell-Cycle-Set.xlsx')
df.dropna(inplace=True)
```

# Week 1

## Task 1

Generate a histogram of one of the cell cycle stages of the RNA and protein distribution. Do you notice anything interesting with regards to the mean/variance of the distribution?

```python
fig1, ax1 = plt.subplots(ncols=3, figsize=(16,4))
df.mean_RNA_G1.hist(ax=ax1[0], label='RNA')
df.mean_protein_G1.hist(ax=ax1[0], label='Protein')
ax1[0].legend()
ax1[0].set_xlabel('Mean G1 Expression')

df.mean_RNA_S.hist(ax=ax1[1], label='RNA')
df.mean_protein_S.hist(ax=ax1[1], label='Protein')
ax1[1].legend()
ax1[1].set_xlabel('Mean S Expression')

df.mean_RNA_G2.hist(ax=ax1[2], label='RNA')
df.mean_protein_G2.hist(ax=ax1[2], label='Protein')
ax1[2].legend()
ax1[2].set_xlabel('Mean G2 Expression')
```

Variance in case of the protein distribution is larger since the histogram obtained is wider than that in case of the RNA. The histogram also indicates that the translation rates are variable. Also, proteins maybe broken down or affected by other process and hence the concentration of proteins is more spread out as compared to RNA, whose main function is translation to proteins. The half-life of different proteins has high variation (minutes to days), whereas the degradation rate of mRNA falls within a narrower range.

## Task 2

Look at the pairwise correlations between each of the RNA/protein columns (this can be achieved using the corr() function). Does the change in timestep have much effect on the relationship(s) between RNA and protein?

```python
fig2, ax2 = plt.subplots(ncols=3, figsize=(12,4))

df.plot.scatter('mean_RNA_G1', 'mean_protein_G1', ax=ax2[0], color='r', label='r={:0.2f}'.format(
            spearmanr(df.mean_RNA_G1.values, df.mean_protein_G1.values)[0]
        ))
x1 = df.iloc[:, 1].values.reshape(-1, 1)
y1 = df.iloc[:, 4].values.reshape(-1, 1)
linear_regressor1 = LinearRegression()
linear_regressor1.fit(x1, y1)
y_pred1 = linear_regressor1.predict(x1)
ax2[0].plot(x1, y_pred1, color='green')

df.plot.scatter('mean_RNA_S', 'mean_protein_S', ax=ax2[1], color='g', label='r={:0.2f}'.format(
            spearmanr(df.mean_RNA_S.values, df.mean_protein_S.values)[0]
        ))
x2 = df.iloc[:, 2].values.reshape(-1, 1)
y2 = df.iloc[:, 5].values.reshape(-1, 1)
linear_regressor2 = LinearRegression()
linear_regressor2.fit(x2, y2)
y_pred2 = linear_regressor2.predict(x2)
ax2[1].plot(x2, y_pred2, color='blue')

df.plot.scatter('mean_RNA_G2', 'mean_protein_G2', ax=ax2[2], color='b', label='r={:0.2f}'.format(
            spearmanr(df.mean_RNA_G2.values, df.mean_protein_G2.values)[0]
        ))
x3 = df.iloc[:, 3].values.reshape(-1, 1)
y3 = df.iloc[:, 6].values.reshape(-1, 1)
linear_regressor3 = LinearRegression()
linear_regressor3.fit(x3, y3)
y_pred3 = linear_regressor3.predict(x3)
ax2[2].plot(x3, y_pred3, color='red')
```

No, the change in timestep has minimal effect on the relationship between RNA and protein. The values of the Spearman correlations for all three phases (Gap 1, Synthesis and Gap 2) in the interphase are close, as the value of the timestep increases by 0.01 from phase to phase.

## Task 3 

Generate a scatterplot of the RNA versus. protein for each cell cycle stage. Fit a linear model to the data, can we infer protein concentration from RNA concentration?

```python
fig3,ax3 = plt.subplots()
cax = ax3.imshow(df.corr(), cmap='magma')
ax3.set_xticklabels(df.columns, rotation=90)
ax3.set_yticklabels(df.columns)
fig3.colorbar(cax)
print(df.corr())
```

No it is not possible to infer protein concentration from RNA concentration. As can be seen from the graph, the correlation coefficient is high (approximately 1) when comparing RNA-RNA concentrations and protein-protein concentrations. But when comparing RNA and protein concentrations, the correlation coefficient is low, which implies that it is difficult to infer protein concentration from RNA concentration.

## Task 4

Calculate the change in mRNA/protein level across the cell cycle by taking the difference at each stage (G1-S, S-G2, G2-G1), and standardize the differences by mean-centering and variance scaling. Repeat tasks 1 and 2 by plotting the changes in levels with GOBP/GOCC labelling. What do we notice about changes in the cell cycle? Is there any apparent clustering of GO terms?

```python
df['mean_RNA_g1s'] = (df.mean_RNA_S - df.mean_RNA_G1)
df['mean_RNA_sg2'] = (df.mean_RNA_G2 - df.mean_RNA_S)
df['mean_RNA_g2g1'] = (df.mean_RNA_G1 - df.mean_RNA_G2)
df['mean_protein_g1s'] = (df.mean_protein_S - df.mean_protein_G1)
df['mean_protein_sg2'] = (df.mean_protein_G2 - df.mean_protein_S)
df['mean_protein_g2g1'] = (df.mean_protein_G1 - df.mean_protein_G2)

# standardise
df.iloc[:,-6:] = (df.iloc[:,-6:] - df.iloc[:,-6:].mean(axis=0)) / df.iloc[:,-6:].std(axis=0)

gobp = df[df.GOBP.str.contains('cell cycle')]
gocc = df[df.GOCC.str.contains('ribosome')]
fig6,ax6 = plt.subplots(ncols=3, figsize=(12,4))
df.plot.scatter('mean_RNA_g1s', 'mean_protein_g1s', ax=ax6[0], color='y', alpha=.5)
df.plot.scatter('mean_RNA_sg2', 'mean_protein_sg2', ax=ax6[1], color='y',alpha=.5)
df.plot.scatter('mean_RNA_g2g1', 'mean_protein_g2g1', ax=ax6[2], color='y', alpha=.5)
ax6[0].scatter(gobp.mean_RNA_g1s, gobp.mean_protein_g1s, color='r', s=10., alpha=.7)
ax6[1].scatter(gobp.mean_RNA_sg2, gobp.mean_protein_sg2, color='r', s=10., alpha=.7)
ax6[2].scatter(gobp.mean_RNA_g2g1, gobp.mean_protein_g2g1, color='r', s=10., alpha=.7)
ax6[0].scatter(gocc.mean_RNA_g1s, gocc.mean_protein_g1s, color='b', s=10., alpha=.7)
ax6[1].scatter(gocc.mean_RNA_sg2, gocc.mean_protein_sg2, color='b', s=10., alpha=.7)
ax6[2].scatter(gocc.mean_RNA_g2g1, gocc.mean_protein_g2g1, color='b', s=10., alpha=.7)
```

There isn't much change between the stages in the cell cycle, except for a few anomalous points. There are a few outlying points, but in general most of them are clustered close to the centre.
The clustering in case of the GO terms is apparent. The clustering of GOBP terms is quite dense. The clustering of GOCC terms is even denser than that of the GOBP terms.
